```python
files = open('artikel_list.txt', 'r')
files.readline()
for index,file in enumerate(files):
    # Download the HTML data from website: https://www.halodoc.com/artikel
    basic_url = file.split(';')[0]
    print(basic_url)
    response = requests.get(basic_url)
    
    # Parse the HTML data and create a BeautifulSoup object.
    soup = BeautifulSoup(response.content, 'html.parser')
    judul_artikel = soup.find("h3", class_="section-header__content-text-title section-header__content-text-title--xlarge").get_text()
    try:
        summary_makanan = soup.find('p',class_='article-page__summary').get_text()
    except:
        summary_makanan = ' '
    gambar_artikel = soup.find('img', class_='default-image-style d-none').get('src')
    body_artikel = soup.find('div',attrs={'class':['article__content'],'style':False}).get_text()
    author = soup.find('a','hd-base-links').get_text().strip()
    createdAt = soup.find('div','article-page__reviewer').get_text().split('\xa0')[2].strip()
    
    dataframe.loc[index,'id'] = str(uuid.uuid1())
    dataframe.loc[index,'title'] = judul_artikel.strip()
    dataframe.loc[index,'summary'] = summary_makanan.strip()
    dataframe.loc[index,'imgUrl'] = gambar_artikel
    dataframe.loc[index,'content'] = body_artikel.strip()
    dataframe.loc[index,'author'] = author.strip()
    dataframe.loc[index,'createdAt'] = createdAt.strip()
    dataframe.loc[index,'tags'] = file.split(';')[1].strip()
    dataframe.loc[index,'link'] = basic_url
```

```python
dataframe.viewers = 0
dataframe.likes = 0
```

```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
```

```python
# Download the HTML data from the website.
basic_url = 'https://mobile.fatsecret.co.id/kalori-gizi'
response = requests.get(basic_url)

# Parse the HTML data and create a BeautifulSoup object.
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
makanan_umum = soup.find_all("div", class_="next-link")
```

```python
column_names = ['Nama','Ukuran Porsi','Energi','',
 'Gula',
 'Kalium',
 'Karbohidrat',
 'Kolesterol',
 'Lemak',
 'Lemak Jenuh',
 'Lemak Trans',
 'Lemak tak Jenuh Ganda',
 'Lemak tak Jenuh Tunggal',
 'Protein',
 'Serat',
 'Sodium']
dataframe = pd.DataFrame(columns=column_names)
dataframe.loc[0]= 0
dataframe
```

```python
makanan_umum = soup.find_all("div", class_="next-link")
index = -1

for i in makanan_umum:
    print('BAB '+i.find('a').get('href'))
    clicked_url_1 = i.find('a').get('href')
    response = requests.get(basic_url+str(clicked_url_1))
    soup = BeautifulSoup(response.content, 'html.parser')
    each_makanan = soup.find_all("div", class_="next-link")[:-18]
    for makanan in each_makanan:
        print('Makanan ',index,makanan.find('a').get('href'))
        clicked_url_2 = makanan.find('a').get('href')
        response = requests.get(basic_url+str(clicked_url_2))
        soup = BeautifulSoup(response.content, 'html.parser')
        
        porsi_makanan = soup.find_all('div',"other-link")[::2]
        for porsi in porsi_makanan:
            index +=1
            print('Porsi ',porsi.find('a').get('href'))
            clicked_url_3 = porsi.find('a').get('href')
            response = requests.get(basic_url+str(clicked_url_3))
            soup = BeautifulSoup(response.content, 'html.parser')
        # informasi_makanan = soup.find_all('div',attrs={'class':['padding page-title',"serving_size black serving_size_label", 'serving_size black serving_size_value','nutrient black left', 'nutrient right tRight', 'nutrient black right tRight','nutrient sub left','nutrient left', 'nutrient black right tRight',]})
        # informasi_makanan = soup.find_all('div',attrs={'class':['padding page-title',"serving_size black serving_size_label", 'serving_size black serving_size_value','nutrient black left', 'nutrient sub left','nutrient left',]})
            tes_title = soup.find('div',attrs={'class':['padding page-title']})
            tes_right = soup.find_all('div',attrs={'class':['serving_size black serving_size_value','nutrient right tRight', 'nutrient black right tRight', 'nutrient black right tRight',]})
            tes_left = soup.find_all('div',attrs={'class':["serving_size black serving_size_label",'nutrient black left','nutrient sub left','nutrient left',]})
            dataframe.loc[index,'Nama'] = re.sub(r'\n','',tes_title.get_text())
            for k,v,in zip(tes_left,tes_right):
                dataframe.loc[index,k.get_text()] = v.get_text()
            # list_kandungan = []
        # for i,v in enumerate(informasi_makanan):
            # if i == 0:
                # continue
            # list_kandungan.append(re.sub(r'\n','',i.get_text()))    
            # print(v.get_text())
            # column_names.add(v.get_text())
            # list_informasi.append(re.sub(r'\n','',v.get_text()))    
        # list_informasi.append(list_kandungan)
```

```python
import re

def clean_df(x):
    # print(type(x))
    if type(x) == str:
        x = x.replace(',','.')
        x = re.sub('[^0-9.]','',x)
    return float(x)
```

```python
cols = dataframe.iloc[:,3:].columns
for col in cols:
    dataframe[col] = dataframe[col].apply(clean_df)
```
